Since there is no index we cant prove mutable or immutable manually so we will use Built_in_Functions:

**Built_in_Functions_set**

`Increasing Memory `
- `add`
- `update `

`Decreasing Memory`
- `pop `
- `discard `
- `remove `

`Some other Tasks `

- `copy `
- `clear `
- `union `
- `intersection `

```python
# add = this function will take single input from user and add value randomlly in the set

a = {10,20,30} # save in memory

a.add(100) # here value will be added randomly

print(a)
```

```python
# update = this function will take multiple input from user and add value randomlly in the set

a = {10,20,30,40,50} # save in Memory

a.update([500,600,700]) # 3 values will be added randomlly into existing set

print(a)
```

```python
# pop = randomly one value will be removed:

a = {50, 20, 500, 40, 10, 600, 700, 30}

a.pop()

print(a)
```

```python
# discard = will take a value from user and if the value is available it will remove
# if the value is not available it will be silent

a = {50, 20, 500, 40, 10, 600, 700, 30}

a.discard(500)

print(a)
```

```python
# discard = will take a value from user and if the value is available it will remove
# if the value is not available it will be silent

a = {50, 20, 500, 40, 10, 600, 700, 30}

a.discard(2000)

print(a)
```

```python
# remove = will take a value from user and if the value is available it will remove
# if the value is not available it will through Error

a = {50, 20, 500, 40, 10, 600, 700, 30}

a.remove(500)

print(a)
```

```python
# remove = will take a value from user and if the value is available it will remove
# if the value is not available it will through Error

a = {50, 20, 500, 40, 10, 600, 700, 30}

a.remove(2000)

print(a)
```

**from the Above Operations we can say Python set is Mutable**

```python
# copy

a = {10,20,30,40}
b = a # deep copy

print(a)
print(b)

print(id(a))
print(id(b))

a.add(500)

print(a)
print(b)
```

```python
# copy

a = {10,20,30,40}
b = a.copy() # shallow copy

print(a)
print(b)

print(id(a))
print(id(b))

a.add(500)

print(a)
print(b)
```

```python
# clear

a = [10,20,30,50]

a.clear()

print(a)
```

```python
# clear

a = {10,20,30,50}

a.clear()

print(a)
```

```python
a = []

a.append(100)
a.append(200)

print(a)
```

```python
a = set()

a.add(100)
a.add(200)

print(a)
```

```python
# union

a = {1,2,3,4,5,6}
b = {5,6,7,8,9,10}

print(a.union(b))
```

```python
# intersection

a = {1,2,3,4,5,6}
b = {5,6,7,8,9,10}

print(a.intersection(b))
```

```python
a = {10,20,30,40,50}

b = list(a)

b.append(100)

c = set(b)
print(c)
```
