```python
# array = which is used to create data in the form of dimension
# 1D

a = np.array([10,20,30,40,50]) # 1D

print(a) # calling from memory

print(type(a)) # checking dtype

print(a.ndim) # used to check number of dims

print(a.shape) # check rows and columns -> (row,column)

print(a.size) # total number of values
```

```python
print(a[2])
```

```python
# 2D

a = np.array([ [1,4,9] , [10,7,3] , [11,8,14]   ] )

print(a) # calling from memory

print(type(a)) # checking dtype

print(a.ndim) # used to check number of dims

print(a.shape) # check rows and columns -> (row,column)

print(a.size) # total number of values
```

```python
print(a[1][2])
```

```python
# arange = arange function will take starting_point and ending_point:
# arange function will give outcome in 1D format


a = np.arange(11,17)

print(a)

print(a.ndim)
```

```python
# 1D to 2D

a = np.arange(1,61)

print(a)
print(a.ndim)

b = a.reshape(10,6)

print(b)
print(b.ndim)
```

```python
a = np.arange(1,61).reshape(10,6)
print(a)
```

```python
print(a[4][2])
```

```python
a = np.arange(1,61).reshape(2,5,6)
print(a)
```

```python
# reshape(-1,1)

a = np.array([10,20,50,90,11,5])
print(a)
print(a.ndim)

b = a.reshape(-1,1)
print(b)
print(b.ndim)
print(b.shape)
```

```python
# reshape(1,-1)

a = np.array([10,20,50,90,11,5])
print(a)
print(a.ndim)

b = a.reshape(1,-1)
print(b)
print(b.ndim)
print(b.shape)
```

```python
# ravel -> any dim will be converted back to 1D

a = np.array([ [1,4,9] , [10,7,3] , [11,8,14]   ] )
print(a)
print(a.ndim)

b = a.ravel()
print(b)
print(b.ndim)
```

```python
# transpose

a = np.array([ [1,4,9] , [10,7,3] , [11,8,14]   ] )

print(a)

b = a.T

print(b)
```

```python
# zeros
# default data type for numpy is float64

a = np.zeros((5,4))
print(a)
print(a.dtype)
```

```python
# zeros
# default data type for numpy is float64

a = np.zeros((5,4) , dtype = np.int64)
print(a)
print(a.dtype)
```

```python
# ones

a = np.ones((5,4),dtype=np.int64)
print(a)
```

```python
# full

a = np.full((5,4) , 9)
print(a)
```

```python
# eye
a = np.eye(4)
print(a)
```

```python
import numpy as np
a = np.array([[1,4,9] , [3,7,2] , [10,18,25]])
d = []
for i in a:
  c = 0
  for j in i:
    c = c + j
  d.append(c)

p = 0
for k in d:
  p = p + k

print(p)
```

```python
import numpy as np
a = np.array([[1,4,9] , [3,7,2] , [10,18,25]])

f_c = []
s_c = []
t_c = []


for i in a:
  f_c.append(i[0])
  s_c.append(i[1])
  t_c.append(i[2])


s_f_c = 0
s_s_c = 0
s_t_c = 0

for j in range(3):
  s_f_c = s_f_c + f_c[j]
  s_s_c = s_s_c + s_c[j]
  s_t_c = s_t_c + t_c[j]

print(s_f_c + s_s_c + s_t_c)
```

```python
import numpy as np
a = np.array([[1,4,9] , [3,7,2] , [10,18,25]])

c = 0

for i in a:
  for j in i:
    if j % 2 == 0:
      c = c + j
print(c)
```
