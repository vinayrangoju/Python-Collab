**Python Dictionary is Mutable or Immutable -> Prove it**


```python
# try to increase the Memory

fruits = {'apple':100,'Banana':150,'carrot':200}

print(fruits)
```

```python
fruits = {'apple':100,'Banana':150,'carrot':200}
print(f'Before : {fruits}')

fruits['apple'] = 500


print(f'After : {fruits}')
```

```python
fruits = {'apple':100,'Banana':150,'carrot':200}
print(f'Before : {fruits}')

fruits['grapes'] = 500


print(f'After : {fruits}')
```

```python
fruits = {} # empty dict

fruits['apple'] = 200
fruits['grapes']  = 500


print(fruits)
```

```python
# trying to delete the Memory

fruits = {'apple':100,'Banana':150,'carrot':200}
print(f'Before : {fruits}')

del fruits['carrot']

print(f'After : {fruits}')
```

**From the Above Operations we can say that Python Dict is Mutable**

**built_in_functions in Dictionary**

- `get`
- `update `
- `pop `
- `popitem `
- `keys `
- `values `
-`items `
- `clear `
- `copy `

```python
fruits = {'apple':100,'Banana':150,'carrot':200}

print(fruits['Banana'])
```

```python
# get

fruits = {'apple':100,'Banana':150,'carrot':200}

print(fruits.get('Banana'))
```

```python
fruits = {'apple':100,'Banana':150,'carrot':200}
print(f'Before : {fruits}')

fruits['grapes'] = 500


print(f'After : {fruits}')
```

```python
# update

fruits = {'apple':100,'Banana':150,'carrot':200}
print(f'Before : {fruits}')

fruits.update({'grapes':500})


print(f'After : {fruits}')
```

```python
# pop = will take column name and removes it

fruits = {'apple': 100, 'Banana': 150, 'carrot': 200, 'grapes': 500}

print(f'Before : {fruits}')

fruits.pop('Banana')


print(f'After : {fruits}')
```

```python
# popitem = will remove last column

fruits = {'apple': 100, 'Banana': 150, 'carrot': 200, 'grapes': 500}

print(f'Before : {fruits}')

fruits.popitem()


print(f'After : {fruits}')
```

```python
# keys

fruits = {'apple': 100, 'Banana': 150, 'carrot': 200, 'grapes': 500}

print(fruits.keys())
```

```python
# values

fruits = {'apple': 100, 'Banana': 150, 'carrot': 200, 'grapes': 500}

print(fruits.values())
```

```python
# items = both key_value pair

fruits = {'apple': 100, 'Banana': 150, 'carrot': 200, 'grapes': 500}

print(fruits.items())
```

```python
# clear = will remove data inside the column but column will be alive

fruits = {'apple': 100, 'Banana': 150, 'carrot': 200, 'grapes': 500}

fruits.clear()

print(fruits)
```

```python
# copy

fruits = {'apple': 100, 'Banana': 150}
b = fruits # deep copy

print(fruits)
print(b)

print(id(fruits))
print(id(b))

fruits['grapes'] = 1000

print(fruits)
print(b)
```

```python
# copy

fruits = {'apple': 100, 'Banana': 150}
b = fruits.copy() # shallow copy

print(fruits)
print(b)

print(id(fruits))
print(id(b))

fruits['grapes'] = 1000

print(fruits)
print(b)
```
