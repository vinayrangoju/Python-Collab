**Python_Tuples**

- `Python Tuple is going to represent with -> ()`
- `Python Tuple can accept any data type `

```python
a = (10,12.55,'data',False) # save in Memory

print(a) # calling from memory
```

```python
print(a[1])
```

```python
print(a[2])
```

```python
print(a[2][2])
```

```python
a = (10,20,30,40,50)

print(a[2:5])
```

```python
print(a[1:])
```

```python
print(a[:4])
```

```python
print(a[::-1])
```

```python
a = (10,20,30,40,50)
print(a[::2])
```

**Tuple is Mutable or Immutable -> prove it**

```python
# Trying to increase the Memory

a = (10,50,12) # save in Memory

a[3] = 100

print(a)
```

```python
# Trying to Decrease the Memory

a = (10,50,12) # save in Memory

del a[1]

print(a)
```

From Above Operations we can say that Python Tuple is Immutable

**Built_in_Functions**

In tuple we have only 2 Built_in_Functions

- `count `
- `index `

```python
# count

a = (10,20,30,40,10,20)
print(a.count(10))
```

```python
# index

a = (10,20,30,40,50)
print(a.index(30))
```
