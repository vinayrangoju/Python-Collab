**String is Mutable or Immutable -> Prove it**

```python
# Mutable = if user is capable of increasing or decreasing the Memory size then it is a mutable
# IMMutable = if user is not capable of increasing or decreasing the Memory size then it is a immutable
```

```python
# Trying to increase the Memory

a = 'data' # save in the Memory

a[4] = 'j'

print(a)
```

```python
# Trying to reduce the Memory size

a = 'data' # save in memory

del a[2]

print(a)
```

**From Above OPerations we can say that String is a Immutable**
