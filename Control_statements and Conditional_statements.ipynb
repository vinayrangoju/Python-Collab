```python
# find the given number is a perfect number or not:

n = 6
c = 0

for i in range(1,n):
  if n % i == 0:
    c = c + i

if n == c:
  print(f'{n} is a Perfect Number')
else:
  print(f'{n} is not a Perfect Number')
```

```python
# dictionary

fruits = {'apple':180,'banana':160,'carrot':400,'grapes':500}
# in dict forloop will collect only keys

for i in fruits:
  if fruits[i] > 200:
    fruits[i] = fruits[i] * 2
  else:
    fruits[i] = fruits[i] * 3

print(fruits)
```
