**Built_in_Functions**

**Increase the Memory**
- `append`
- `extend `
- `insert `

**Decrease the Memory**
- `pop `
- `remove `

**Do some Specific Task**
- `copy `
- `clear `
- `count `
- `index `
- `sort `
- `reverse `

```python
# append = this function will take only one input from the user and will add to the end of the existing List

a = [10,4,24,19,50,33,89] # save in Memory

a.append(100)

print(a)
```

```python
# extend = this function will take multiple inputs from user and add to the end of existing list:

a = [10,4,24,19,50,33,89] # save in Memory

a.extend([100,200,300])

print(a)
```

```python
# insert = we can add value where we want in the existing List:

a = [10,4,24,19,50,33,89] # save in Memory

a.insert(3,100)

print(a)
```

```python
# pop = defaulty will remove last value
# but if we pass a value it will treat like index

a = [10,4,24,19,50,33,89] # save in Memory

a.pop()

print(a)
```

```python
# pop = defaulty will remove last value
# but if we pass a value it will treat like index

a = [10,4,24,19,50,33,89] # save in Memory

a.pop(3)

print(a)
```

```python
# remove = give value it will be removed

a = [10,4,24,19,50,33,89] # save in Memory

a.remove(50)

print(a)
```

**From Above Operations we can say that Python List is Mutable**

```python
# count:

a = [10,4,24,19,50,33,89] # save in Memory

print(a.count(24))
```

```python
# index

a = [10,4,24,19,50,33,89] # save in Memory

print(a.index(24))
```

```python
# reverse = is same like [::-1]

a = [10,4,24,19,50,33,89] # save in Memory

print(a[::-1])
```

```python
# reverse = is same like [::-1]

a = [10,4,24,19,50,33,89] # save in Memory

a.reverse()

print(a)
```

```python
# sort => ascending order and descending order:
# defaulty sort will apply ascending order

a = [10,4,24,19,50,33,89] # save in Memory

a.sort()

print(a)
```

```python
# sort => ascending order and descending order:
# defaulty sort will apply ascending order

a = [10,4,24,19,50,33,89] # save in Memory

a.sort(reverse=True)

print(a)
```

```python
a = [10,4,24,19,50,33,89] # save in Memory

print(a.index(24))

a.sort()

print(a.index(24))
```

```python
a = [10,20,30] # save in the Memory

del a

print(a) # calling from memory
```

```python
# clear = will not delete varibale will only remove data in the varibale

a = [10,20,30]  #save in Memory

a.clear()

print(a) # calling from Memory
```

```python
# copy

a = [10,20,30,40]
b = a   # deep copy

print(a)
print(b)

print(id(a))
print(id(b))

a.append(100)

print(a)
print(b)
```

```python
# copy

a = [10,20,30,40]
b = a.copy()   # shallow copy

print(a)
print(b)

print(id(a))
print(id(b))

a.append(100)

print(a)
print(b)
```
