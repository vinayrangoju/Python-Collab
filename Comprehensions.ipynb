**Comprehensions**

Comprehensions are used to reduce the Lines of code:

There are 2 types of comprehensions:

- `List Comprehensions`
- `Dictionary Comprehensions`

**List Comprehensions**

- `syntax : 1 -> forloop and print `

        - [<print_area> <forloop_area>]

- `syntax : 2 -> forloop and if and print `

        - [<print_area> <forloop_area> <if_area>]
- `syntax : 3 -> forloop and if else and print `


        - [<if_related_print> <if_area> <else_area> <else_print> <forloop_area>]

```python
for i in range(1,11):
  print(f'Good Morning')
```

```python
print(['Good Morning' for i in range(1,11)])
```

```python
for i in range(11,21):
  if i % 2 == 0:
    print(f'Even : {i}')
```

```python
print([f'Even : {i}' for i in range(11,21) if i % 2 == 0])
```

```python
for i in range(11,21):
  if i % 2 == 0:
    print(f'Even : {i}')
  else:
    print(f'Odd : {i}')
```

```python
print([f'Even : {i}' if i % 2 == 0 else f'Odd : {i}' for i in range(11,21)])
```

**Dictionary comprehension**

- `syntax : 1 -> forloop and print `

        - {<print_area> <forloop_area>}

- `syntax : 2 -> forloop and if and print `

        - {<print_area> <forloop_area> <if_area>}
- `syntax : 3 -> forloop and if else and print `


        - {<if_related_print> <if_area> <else_area> <else_print> <forloop_area>}

```python
fruits = {'apple':20,'banana':30,'carrot':50}

for i in fruits:
  fruits[i] = fruits[i] + 20

print(fruits)
```

```python
fruits = {'apple':20,'banana':30,'carrot':50}

print({i:fruits[i]+20 for i in fruits})
```

```python
fruits = {'apple':20,'banana':30,'carrot':50}
d = {}

for i in fruits:
  if fruits[i] > 25:
    d[i] = fruits[i]

print(d)
```

```python
fruits = {'apple':20,'banana':30,'carrot':50}

print({i:fruits[i] for i in fruits if fruits[i] > 25})
```

```python
fruits = {'apple':20,'banana':30,'carrot':50}


for i in fruits:
  if fruits[i] > 25:
    fruits[i] = fruits[i] * 3
  else:
    fruits[i] = fruits[i] * 2


print(fruits)
```

```python
fruits = {'apple':20,'banana':30,'carrot':50}

print(fruits.items())
```

```python
fruits = {'apple':20,'banana':30,'carrot':50}

print({i:(j * 3 if j > 25 else j * 2) for i,j in fruits.items()})
```

```python
# without comprehesion:

import time
starting_time = time.time()

fruits = {'apple':20,'banana':30,'carrot':50}


for i in fruits:
  if fruits[i] > 25:
    fruits[i] = fruits[i] * 3
  else:
    fruits[i] = fruits[i] * 2


print(fruits)

print(f'Total time to run the code : {time.time() - starting_time}')
```

```python
# with comprehension

# without comprehesion:

import time
starting_time = time.time()

fruits = {'apple':20,'banana':30,'carrot':50}

print({i:(j * 3 if j > 25 else j * 2) for i,j in fruits.items()})

print(f'Total time to run the code : {time.time() - starting_time}')
```
