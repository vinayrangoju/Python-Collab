**Lambda Function**

- `Lambda function is also known as ananymous Function `
- `ananymous Function -> A function without a function name `
- `syntax -> lambda arguments : Expressions `


```python
def fun():
  print('Good Morning')


fun() # calling
```

```python
c = lambda x,y : x + y


print(c(10,20)) # calling lambda function
```

```python
def fun(n):
  return n * n


sol = fun(5) # calling
print(sol)
```

```python
c = lambda a : a * a

print(c(5)) # calling
```

```python
n = 2

for i in range(1,11):
  print(n,'*',i,'=',n*i)
```

```python
# normal

n = 5

for i in range(1,11):
  print(f'{n} * {i} = {n*i}')
```

```python
# function

def sharuk(n):
  for i in range(1,11):
    print(f'{n} * {i} = {n*i}')


sharuk(2) # calling
```

```python
sharuk(5)
```

```python
# Lambda Function:

c = lambda x,y : x * y
n = 2

for i in range(1,11):
  print(f'{n} * {i} = {c(n,i)}')
```

```python
# Map Function:

# Map Syntax : map(function,iterables)

def kumar(b):
  if b % 2 == 0:
    return b
  else:
    return "Not Found"


sol = map(kumar,[10,21,19,22,43,84])

print(list(sol))
```

```python
sol = map(lambda x : x % 2 == 0,[10,21,19,22,43,84])

print(list(sol))
```

```python
# Map Function:

# Map Syntax : map(function,iterables)

def kumar(b):
  if b % 2 == 0:
    return b


sol = map(kumar,[10,21,19,22,43,84])

print(list(sol))
```

```python
# filter:

def kumar(b):
  if b % 2 == 0:
    return b


sol = filter(kumar,[10,21,19,22,43,84])

print(list(sol))
```

```python
sol = filter(lambda x : x % 2 == 0,[10,21,19,22,43,84])

print(list(sol))
```

```python
# reduce

import functools
print(functools.reduce(lambda x,y : x+y , [1,4,8,9,11,18]))
```

**Map with Filter**

```python
a=map(lambda x:x + 10,filter(lambda x:x%2==0,[10,11,12,14,15,19,22]))
print(list(a))
```

**Filter with Map**

```python
a=filter(lambda x:x%2==0,map(lambda x:x+5,[10,11,12,14,17,20,22,21,25]))
print(list(a))
```

**reduce with Map and filter**

```python
a=functools.reduce(lambda x,y :x + y,map(lambda x:x + 10,filter(lambda x:x%2==0,[10,11,12,14,15,19,22])))
print(a)
```
