**variance and std**

```python
df = pd.DataFrame({'kohli':[28,34,32,30,36,29],
                   'rohith':[18,81,67,0,0,43]})
```

```python
df
```

```python
print(f"Mean of kohli : {df['kohli'].mean()}")
print(f"Mean of Rohith : {df['rohith'].mean()}")
```

```python
print(f"Variance of kohli : {df['kohli'].var()}")
print(f"Variance of Rohith : {df['rohith'].var()}")
```

```python
print(f"Standard Deviation of kohli : {df['kohli'].std()}")
print(f"Standard Deviation of Rohith : {df['rohith'].std()}")
```
