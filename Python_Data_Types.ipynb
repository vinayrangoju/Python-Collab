**Rules to create a varibales**

- `Rule_no:1 -> Varibales should not start with a number if you want to start please use underscore `

```python
10a = 20
print(10a)
```

```python
_10a = 20
print(_10a)
```

- `Rule_no_2 : For varibales use either single character or single word as a variable -> multiple words are not allowed as a varibale if you need please use underscore`

```python
kumar = 20
print(kumar)
```

```python
sai kumar = 20
print(sai kumar)
```

```python
sai_kumar = 20
print(sai_kumar)
```

- `Rule_no 3-> Varibales are case sensitive -> Capital letters means capital small letters means small `

```python
city = 'usa' # save in the Memory
print(CITY) # calling from Memory
```

- `Rule_no 4: Dont use below punctuations as a varibale -> !"#$%&'()*+,-./:;<=>?@[\]^_`{|}~`

```python
& = 100
print(&)
```

**Type_Conversions**

- `int -> float and int -> str `
- `float -> int and float -> str `
- `str -> int and str -> float `

```python
# int -> float

a = 12 # Save in Memory
print(a) # calling from Memory
print(type(a)) # checking data type of a varibale


b = float(a) # actual conversion is going on here
print(b) # calling from Memory
print(type(b)) # checking data type of a varibale
```
