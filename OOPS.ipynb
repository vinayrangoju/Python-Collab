Constructor is just like class name only

There are 2 types of constructors

- `Non_Parameter constructor `
- `Parameter constructor `

```python
# Non Parameter constructor:

class CSE:
  def __init__(self):  # self -> object copy | reference
    print('good morning')


kumar = CSE() # constructor will be called
```

```python
class CSE:
  def __init__(self,a,b): # self is just a varibale which holds object copy
    self.a = a
    self.b = b

kumar = CSE(10,20) # constructor called

print(kumar.a + kumar.b)
```

```python
class MECH:

  def addition(self):
    p = 50
    g = 80
    print(p+g)

  def subtraction(self):
    p = 50
    g = 80
    print(p-g)


kumar = MECH()
kumar.addition()
kumar.subtraction()
```

```python
class MECH:
  def __init__(self,p,g):
    self.p = p
    self.g = g

  def addition(self):
    print(self.p + self.g)

  def subtraction(self):
    print(self.p - self.g)

  def common(self):
    self.addition()
    self.subtraction()

kumar = MECH(80,50)
kumar.common()
```

```python
# find the perfect number:

class PER:
  def __init__(self,n):
    self.n = n

  def cal(self):
    self.c = 0
    for i in range(1,self.n):
      if self.n % i == 0:
        self.c = self.c + i

    if self.n == self.c:
      print(f'Perfect : {self.n}')
    else:
      print(f'Not a perfect : {self.n}')

kumar = PER(8)
kumar.cal()
```

**Inheritance**

Inheritance works with parent and child concept:

and they are 5 types of inheritance:

- `single inheritance `
- `multilevel inheritance `
- `multiple inheritance `
- `hierarical inheritance `
- `hybrid inheritance `

```python
# single inheritance:

class PARENT:
  def building(self):
    print('I have a building')


class CHILD(PARENT):
  def job(self):
    print('I have my own job')


anil = PARENT()
anil.building()
```

```python
# single inheritance:

class PARENT:
  def building(self):
    print('I have a building')


class CHILD(PARENT):
  def job(self):
    print('I have my own job')


kumar = CHILD()
kumar.job()
kumar.building()
```

```python
# Multilevel inheritance:

class GRANDPARENT:
  def shopping_mall(self):
    print('I am having shopping mall')


class FATER(GRANDPARENT):
  def building(self):
    print('I have a building')


class CHILD(FATER):
  def job(self):
    print('I have my own job')

prem = GRANDPARENT()
prem.shopping_mall()
```

```python
# Multilevel inheritance:

class GRANDPARENT:
  def shopping_mall(self):
    print('I am having shopping mall')


class FATER(GRANDPARENT):
  def building(self):
    print('I have a building')


class CHILD(FATER):
  def job(self):
    print('I have my own job')

anil = FATER()
anil.building()
anil.shopping_mall()
```

```python
# Multilevel inheritance:

class GRANDPARENT:
  def shopping_mall(self):
    print('I am having shopping mall')


class FATER(GRANDPARENT):
  def building(self):
    print('I have a building')


class CHILD(FATER):
  def job(self):
    print('I have my own job')

kumar = CHILD()
kumar.job()
kumar.building()
kumar.shopping_mall()
```

```python
# Multiple Inheritance:

class MOTHER:
  def shopping_mall(self):
    print('I am having shopping mall')


class FATER:
  def building(self):
    print('I have a building')


class CHILD(FATER,MOTHER):
  def job(self):
    print('I have my own job')

kumar = CHILD()
kumar.job()
kumar.building()
kumar.shopping_mall()
```

```python
# hierarichal inheritance:
class FATHER:
  def shopping_mall(self):
    print('I am having shopping mall')


class B1(FATHER):
  def building(self):
    print('I have a building')


class B2(FATHER):
  def job(self):
    print('I have my own job')

class B3(FATHER):
  def bussiness(self):
    print('I have own bussiness')

anil = B2()
anil.job()
anil.shopping_mall()
```

```python
# hierarichal inheritance:
class FATHER:
  def shopping_mall(self):
    print('I am having shopping mall')


class B1(FATHER):
  def building(self):
    print('I have a building')


class B2(FATHER):
  def job(self):
    print('I have my own job')

class B3(B1,B2):
  def bussiness(self):
    print('I have own bussiness')

anil = B3()
anil.job()
anil.shopping_mall()
anil.building()
anil.bussiness()
```

**polymorphisam**

- `This word dervied from Greek language `

`poly -> Many `

`Morphisam -> forms `

- `They are 2 types of polymorphisam `
       - `Method [Function] overloading `
       - `Method [Function] Overriding `

```python
# method overriding:

def fun():
  print('good morning')

def fun():
  print('good evening')

fun() # calling the function:
```

```python
# Method overriding is only possible with parent and child class

class PARENT:
  def sharuk(self):
    print('I am from india')


class CHILD(PARENT):
  def sharuk(self):
    print('I am from usa ')


anil = CHILD()
anil.sharuk()
```

```python
# method overloading:

def fun(a=None,b=None):
  if a==None and b == None:
    print('I am from india')
  elif a == None and b!=None:
    print('I am from usa ')

  elif a!=None and b==None:
    print('I am from uk')
  else:
    print('I am from canada')


fun()
fun(a = 10)
fun(b=10)
fun(10,20)
```

```python
# abstaction:

from abc import ABC,abstractmethod

class paytm_v1(ABC):
  @abstractmethod
  def bank_transfer(self):
    pass


class patym_v3(paytm_v1):
  def bank_transfer(self):
    print('Amount successfull')

anil = patym_v3()
anil.bank_transfer()
```
