```python
# what is difference between global varibale and local varibale
```

```python
# global varibale = we can use inside the function and outside the function
```

```python
p = 500  # global variable

def fun():
  print(f'I am from inside the function : {p}')

fun() # calling

print(f'I am from outside the function : {p}')
```

```python
def fun():
  m = 900 # local varibale
  print(f'I am from inside the function : {m}')

fun() # calling

print(f'I am from outside the function : {m}')
```

```python
def fun():
  global m
  m = 900 # local varibale
  print(f'I am from inside the function : {m}')

fun() # calling

print(f'I am from outside the function : {m}')
```
